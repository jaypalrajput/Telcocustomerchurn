```python
#Outlier Detection and removal
''' Detection '''
# IQR
Q1 = np.percentile(telcocdf['Tenure Months'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(telcocdf['Tenure Months'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
# Upper bound
upper = np.where(telcocdf['Tenure Months'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(telcocdf['Tenure Months'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
telcocdf.drop(upper[0], inplace = True)
telcocdf.drop(lower[0], inplace = True)
```
